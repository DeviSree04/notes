# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies =['ayalaan','don','maankarathe','marina','seemaraja','VVPS','MKP','kbkr']
percentage =[90,95,85,70,75,100,60,50]
```

# bar chart

```python
plt.bar(movies,percentage,color="yellow",edgecolor="green") #comparing 2 variables
```

# scatter plot

```python
plt.scatter(movies,percentage,color="yellow") #relationship between 2 variables(2 meh ++ strong,1++ 1-- week relation ++-- ilana spread)
```

# line plot

```python
plt.plot(movies,percentage,linestyle='solid') #recent trends
```

# pie chart

```python
plt.pie(percentage,labels=movies) #percentage
```

# Relationship plot

```python
sns.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",
            style="day",kind="scatter") #relation btv variables
```

# Box plot whisker plot

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tip)
```

# Regression plot (prediction)

```python
sns.regplot(x='total_bill',y='tip',data=tip)  #line name line of best fit
```
