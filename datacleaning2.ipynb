# MISSING VALUES

```python
import numpy as np
import pandas as pd
```

```python
data = {'First score':[100,90,np.nan,85],
        'second score':[70,np.nan,80,np.nan],
        'third score':[np.nan,75,np.nan,np.nan]}
   
```

```python
scores = pd.DataFrame(data)
```

```python
scores
```

```python
scores.isnull()
```

```python
scores.isnull().sum()          #to find how many null value
```

```python
#to fill the missing values(inputer-applying)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,
                      strategy='median')
```

```python
imputer.fit(scores)
```

```python
imputer.transform(scores)
```
